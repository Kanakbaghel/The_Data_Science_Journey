```python
import scipy.stats as stats
from math import sqrt,floor, ceil
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mu = 17.5
sigma = 4
n = 50
```

#### a.	Show the sampling distribution of , the sample mean average for a sample of 50 unemployed individuals.

#### for sampling distribution std = sigma/sqrt(n)

```python
np.random.seed(121)
sample = np.random.normal(loc = mu, scale = sigma/sqrt(n), size = 1000)
plt.figure(figsize = (10,5))
sns.displot(data = sample, kind = 'kde',aspect = 2.5, linewidth = 4, color = 'green')
plt.xticks(np.arange(floor(sample.min()), ceil(sample.max()), 0.5))
plt.show()
```

#### b.	What is the probability that a simple random sample of 50 unemployed individuals will provide a sample mean within 1 week of the population mean?

```python
# withinn 1 week of population
x1 = mu - 1
x2 = mu + 1

stats.norm.cdf(x2, loc = mu, scale = sigma/sqrt(n))- stats.norm.cdf(x1, loc = mu, scale = sigma/sqrt(n))
```

```python
# withinn 1/2 week of population
x1 = mu - 1/2
x2 = mu + 1/2

stats.norm.cdf(x2, loc = mu, scale = sigma/sqrt(n))- stats.norm.cdf(x1, loc = mu, scale = sigma/sqrt(n))
```

#### a.	What is the probability a random sample of 90 test takers will provide a sample mean test score within 10 points of the population mean of 502 on the Critical Reading part of the test?

```python
# sampling distribution 
mu_cr = 502
x1_cr = mu_cr -10
x2_cr = mu_cr +10
sigma = 100
n = 90

p = stats.norm.cdf(x2_cr, loc = mu_cr, scale = sigma/sqrt(n)) - stats.norm.cdf(x1_cr, loc = mu_cr, scale = sigma/sqrt(n))
print('Prob within 1 standard deviation = {:.2f}%'.format(p * 100))
```
